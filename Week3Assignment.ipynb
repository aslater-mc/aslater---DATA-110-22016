## Style One - Classic

```python
x = np.linspace(-2, 4, 100)
y = x**2 - 2*x + 1
plt.plot (x, y,)
plt.legend(['f(x)=x2−2x+1'])
plt.style.use('dark_background')
plt.grid()
plt.title('Basic Quadratic Formula Graph')
```

## Style Two - Dark Background

```python
x = np.linspace(-2, 4, 100)
y = x**2 - 2*x + 1
plt.plot (x, y,)
plt.legend(['f(x)=x2−2x+1'])
plt.style.use('seaborn-whitegrid')
plt.grid()
plt.title('Basic Quadratic Formula Graph')
```

## Style Three - Seaborn Whitegrid

```python
x = np.linspace(-2, 4, 100)
y = x**2 - 2*x + 1
plt.plot (x, y,)
plt.legend(['f(x)=x2−2x+1'])
plt.style.use('classic')
plt.grid(c='paleturquoise')
plt.title('Basic Quadratic Formula Graph')
```
