```python
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Task 1: Basic Bar Chart for Crime Rates in 2023

In this task, you will create a bar chart showing the offense rates per 100,000 people for each type of crime in 2023.


Instructions:

Create a bar chart for crime rates in 2023.

Customize the bar colors and add gridlines.

Label the axes and add a title

```python
plt.figure(figsize=(10,10))
x = ['Aggravated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape', 'Robbery',]
x_indexes = np.arange(len(x))
y = [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5]
z = [-3.3, -8.1, -4.9, 12.0, -12.0, -9.8, -0.8]
barplot2 = plt.bar(x_indexes-0.125,y,color='tomato', edgecolor='red',width=0.25, label='Offenses per 100,000 People (2023)')
plt.bar_label(barplot2, labels='z', label_type='edge', color='red', padding=6)
plt.legend()
plt.xticks(ticks=x_indexes, labels=x, rotation=25)
plt.ylabel('Amount of Crimes Committed (2023)', color='maroon')
plt.xlabel('Specific Crimes Committed (2023)', color='maroon')
plt.grid(alpha=0.3, color='skyblue', zorder=0)
plt.tight_layout()
plt.title('Crime Data Visulization 2023')
plt.show()
```

```python
plt.figure(figsize=(10,10))
x = ['Aggravated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape', 'Robbery',]
x_indexes = np.arange(len(x))
y = [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5]
z = [-3.3, -8.1, -4.9, 12.0, -12.0, -9.8, -0.8]
barplot1 = plt.bar(x_indexes+0.125,z,color='skyblue', edgecolor='royalblue', width=0.25, label='Change in Offenses per 100,000 People (2022-2023)')
barplot2 = plt.bar(x_indexes-0.125,y,color='tomato', edgecolor='red',width=0.25, label='Offenses per 100,000 People (2023)')
plt.bar_label(barplot1, labels='y', label_type='edge', color='royalblue', padding=6)
plt.bar_label(barplot2, labels='z', label_type='edge', color='red', padding=6)
plt.legend()
plt.xticks(ticks=x_indexes, labels=x, rotation=45)
plt.yticks()
plt.ylabel('Amount of Crimes Committed (2023)', color='maroon')
plt.xlabel('Specific Crimes Committed (2023)', color='maroon')
plt.grid(zorder=1, alpha=0.3, color='skyblue')
plt.show()
```
